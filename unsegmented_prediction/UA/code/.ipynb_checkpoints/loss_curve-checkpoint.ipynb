```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import MultipleLocator
import os

log_dir = "../model/vnet_supervisedonly_dp/log.txt"
file=open(log_dir)
lines=file.readlines()
rows=len(lines)
lines.pop(0)
lines.pop(0)


loss = []
for line in lines:
    li = line.split(':',5)
    if len(li)<6:
        continue
    else:
        li = li[4]
        li = li.split(',')
        li = li[0]
        li = li.split(' ')
        li = li[1]
        num = float(li)
        loss.append(num)
print(len(loss))

x = [i for i in range(len(loss))]

plt.figure(dpi=300)
plt.plot(x, loss, label="loss")
plt.xlabel("iteration")
plt.ylabel("loss")
'''x_major_locator=MultipleLocator(500)
ax=plt.gca()
ax.xaxis.set_major_locator(x_major_locator)'''
results_dir = "../result"
filename=os.path.join(results_dir,'loss_curve.png')
plt.savefig(filename)
plt.show()
```
